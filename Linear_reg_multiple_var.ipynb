# linear regression using multiple variables

```python
import pandas as pd
import numpy as np
from sklearn import linear_model 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("house_prc.csv")
df
```

```python
import math
x=math.floor(df.bedrooms.median())
x
```

```python
df.bedrooms=df.bedrooms.fillna(x)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["area","bedrooms","age"]],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y=df[["area","bedrooms","age"]]
p=reg.predict(y)
```

```python
df["predict"]=p
df
```

```python
reg.predict([[3000,3,15]])
```

# salary prediction using multivariant linear regression

```python
data=pd.read_csv("hiring.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
med=math.floor(data.test_score.median())
med
```

```python
data.test_score=data.test_score.fillna(med)
data.experience=data.experience.fillna("zero")
# data.experience=data.experience.replace(0,"zero")
data
```

```python
!pip install word2number
```

```python
from word2number import w2n
```

```python
data.experience=data.experience.apply(w2n.word_to_num)
data
```

```python
ref=linear_model.LinearRegression()

ref.fit(data[["experience","test_score","interview_score"]],data["salary($)"])
```

```python
y=data[["experience","test_score","interview_score"]]
f=ref.predict(y)
```

```python
data["pred_salary"]=f
data
```

```python
ref.coef_
```

```python
ref.intercept_
```
